```python
x=new_df.drop('Survived',axis=1)
y=new_df['Survived']
print('feature dimension',x.shape)
print('label dimension',y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif=pd.DataFrame()
vif['VIF values']=[variance_inflation_factor(x.values,i)for i in range(len(x.columns))]
vif['Features']=x.columns
vif
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
```

Now we will move towards creating train test split.

```python
maxacc=0
maxrs=0
for i in range(1,250):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    pred=lr.predict(x_test)
    acc=r2_score(y_test,pred)
    if acc>maxacc:
        maxacc=acc
        maxrs=i
print('max r2 score is :',maxacc,'on random state :',maxrs)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
lr_test=lr.predict(x_test)
lr_train=lr.predict(x_train)
print('r2_score',r2_score(y_test,lr_test))
print('r2 score on training data',r2_score(y_train,lr_train)*100)
print('mean absolute error',mean_absolute_error(y_test,lr_test))
print('mean squared error',mean_squared_error(y_test,lr_test))
print('root mean squared error',np.sqrt(mean_squared_error(y_test,lr_test)))
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=lr_test,y=y_test,color='r')
plt.plot(lr_test,lr_test,color='r')
plt.xlabel('actual',fontsize=14)
plt.ylabel('linear regression',fontsize=14)
plt.title('Linear Regression',fontsize=20)
plt.show()
```
