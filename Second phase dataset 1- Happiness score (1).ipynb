Here, we cannot get the exact idea about outliers. We will try to find out the outliers with the help of zscore method.

```python
from scipy.stats import zscore
outliers=df[['Region','Happiness Rank', 'Dystopia Residual',
       'Standard Error', 'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
       'Generosity','Happiness Score']]
z=np.abs(zscore(outliers))
z
```

```python
np.where(z>3)
```

From the above zscore details we can see that few outliers are present in data.

```python
z.iloc[40,1]
```

```python
df1=df[(z<3).all(axis=1)]
df1.shape
```

We have cleaned the outliers and now we have the new dataframe.

```python
print('Old data shape',df.shape)
print('New data shape',df1.shape)
```

```python
print('Data loss percentage',((df.shape[0]-df1.shape[0])/df.shape[0])*100)
```

The above data loss is absolutely affordable. Now we can check with IQR method.

```python
Q1=outliers.quantile(0.25)
Q3=outliers.quantile(0.75)
IQR=Q3-Q1
df2=df[~((df<(Q1-1.5*IQR)))|(df>(Q3+1.5*IQR)).any(axis=1)]
df2
```

From the above IQR calculations, we can see that there are few outliers still present.

```python
df=new_df.drop('Country',axis=1)
```

```python
x=df.drop('Happiness Score',axis=1)
y=df['Happiness Score']
print('Features Dimension',x.shape)
print('Target Dimension',y.shape)
```

The above code shows the data separation in features and target variable.

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif['VIF Values']=[variance_inflation_factor(x.values,i)for i in range(len(x.columns))]
vif['Features']=x.columns
vif
```

```python
x.drop('Happiness Rank',axis=1,inplace=True)
vif
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
```

Now we will create test trtain split

```python
maxacc=0
maxrs=0
for i in range(1,250):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    pred=lr.predict(x_test)
    acc=r2_score(y_test,pred)
    if acc>maxacc:
        maxacc=acc
        maxrs=i
print('maximum r2 score is :',maxacc,'on random state:',maxrs)

```

```python
LR=LinearRegression()
LR.fit(x_train,y_train)
LR_test=LR.predict(x_test)
LR_train=LR.predict(x_train)
print('r2 score:',r2_score(y_test,LR_test))
print('r2 score on training data:',r2_score(y_train,LR_train)*100)
print('mean absolute error:',mean_absolute_error(y_test,LR_test))
print('mean squared error:',mean_squared_error(y_test,LR_test))
print('root mean sqaured error:',np.sqrt(mean_squared_error(y_test,LR_test)))
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=LR_test,y=y_test,color='r')
plt.plot(LR_test,LR_test,color='r')
plt.xlabel('actual',fontsize=14)
plt.ylabel('linear regression',fontsize=14)
plt.title('Linear Regression',fontsize=20)
plt.show()
```
